```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/drive/My Drive/Data Tugas Statistika/'
beras = pd.read_excel(path + 'Tugas8.xlsx')
```

```
beras.head()
```

```
harga = beras.pivot("Bulan", "Tahun", "Harga")
plt.figure(figsize=(10,10))
cmap = sns.cm.rocket_r
ax = sns.heatmap(harga, 
                 annot=True,
                 fmt="d",
                 cmap = cmap)
plt.show()
```

```
plt.figure(figsize=(15, 5))
sns.set_context("notebook", rc={"lines.linewidth": 2.5})
sns.set_style("whitegrid")
sns.lineplot(data=beras, x="Bulan", y="Harga", hue='Tahun', marker="o", palette="deep")
```
